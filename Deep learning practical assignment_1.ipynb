```python
import numpy as np
import matplotlib.pyplot as plt

from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
def create_dataset(start=1, end=200):
    x = []
    y = []
    for i in range(start, end,2):
        x.append(i)
        y.append(i+2)
    
    return x,y
```

```python
x,y = create_dataset(end = 500)
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
def create_network():
    model = Sequential()
    model.add(Dense(12, input_dim=1, activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='relu'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
model = create_network()
model.fit(x_train, y_train, epochs=400, batch_size=10)
```

<b>We will round the value of predicted values and check the score

```python
def check_score(x,y,model):
    
    X = [round(i) for i in model.predict(x).reshape(1,-1)[0]]
    print("Score is = "+str(accuracy_score(X,y)))
```

```python
#Train score 
check_score(x_train,y_train,model)
```

```python
#Test score
check_score(x_test,y_test, model)
```
